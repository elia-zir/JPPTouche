#  Statistical Analysis

```python
import matplotlib.pyplot as plt
from bioinfokit.analys import stat

import pandas as pd
import numpy as np
import scipy as sp
```

## Import Data
Load the data from a .csv file


```python
filename = "nDCG@5_All_Runs.csv"

#load the data
df = pd.read_csv(filename, sep = ',')

df
```

## BM25 with Synonyms

```python
#Retrieve data
data = df[['Bm25 No Synonyms','Bm25 Syn01','Bm25 Syn02','Bm25 Syn03','Bm25 Syn04','Bm25 Syn05','Bm25 Syn06','Bm25 Syn07','Bm25 Syn08','Bm25 Syn09','Bm25 Syn10']]

#Boxplot
values = data.values
plt.boxplot(values)
plt.ylabel('nDCG@5')
plt.title('BM25 Synonyms')
plt.xticks(np.arange(1,12), ['Bm25 No Synonyms','Bm25 Syn01','Bm25 Syn02','Bm25 Syn03','Bm25 Syn04','Bm25 Syn05','Bm25 Syn06','Bm25 Syn07','Bm25 Syn08','Bm25 Syn09','Bm25 Syn10'] ,rotation=90)
plt.show()

#Reshape dataFrame
df_melted = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['Bm25 No Synonyms','Bm25 Syn01','Bm25 Syn02','Bm25 Syn03','Bm25 Syn04','Bm25 Syn05','Bm25 Syn06','Bm25 Syn07','Bm25 Syn08','Bm25 Syn09','Bm25 Syn10'])
df_melted.columns = ['index','runs','value']

#Perform one-way anova
res = stat()
res.anova_stat(df=df_melted, res_var='value', anova_model='value ~ runs')

print("\n \n ************** ONE WAY ANOVA **************** \n")
print(res.anova_summary)

#Perform multiple pairwise comparison (Tukey's HSD)
res = stat()
res.tukey_hsd(df=df_melted, res_var='value', xfac_var='runs', anova_model='value ~ runs')

print("\n \n ************** PAIRWISE T TEST **************** \n")
print(res.tukey_summary)
```

In Fig.x, we have the boxplots of the runs obtained by using BM2 5and query expansions with synonyms taken from Wordnet (Tab. x). By looking at the plot we can notice that most runs seem to have a normal distribution, while the last run presents an outlier. Furthermore, we can observe that the first seven runs look similar among them, while the last four can beconsidered as a part of a second group.

We perform ANOVA Test (Tab. x) on the runs and we obtained a p-value of 0.000405. Since the p-value is lower than 0.05 we reject H0 (the null hypothesis that all the means are equal) meaning that there is at least one pair of runs that are significantly different. In fact, performing a multiple pairwise comparison (Tukeys HSD Test) (Tab. x), we can notice that runs such as BM25 Syn01 and BM25 Syn10 are significantly different due to a p-value (0.015635) below 0.05. Since the use of query expansion with synonyms, does not leads to an highly significant difference (only 4 pairs are significantly different), it is better to avoid overhead by not includin query expansion. Hence, we can confirm that synonyms do not work well and do not increasethe performance as we stated in section X.X

## LMDirichlet with Synonyms

```python
data = df[['Dirichlet No Synonyms','Dirichlet Syn01','Dirichlet Syn02','Dirichlet Syn03','Dirichlet Syn04','Dirichlet Syn05','Dirichlet Syn06','Dirichlet Syn07','Dirichlet Syn08','Dirichlet Syn09','Dirichlet Syn10']]

#Boxplot
values = data.values
plt.boxplot(values)
plt.ylabel('nDCG@5')
plt.title('LMDirichlet Synonyms')
plt.xticks(np.arange(1,12), ['Dirichlet No Synonyms','Dirichlet Syn01','Dirichlet Syn02','Dirichlet Syn03','Dirichlet Syn04','Dirichlet Syn05','Dirichlet Syn06','Dirichlet Syn07','Dirichlet Syn08','Dirichlet Syn09','Dirichlet Syn10'] ,rotation=90)
plt.show()

#Reshape dataFrame
df_melted = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['Dirichlet No Synonyms','Dirichlet Syn01','Dirichlet Syn02','Dirichlet Syn03','Dirichlet Syn04','Dirichlet Syn05','Dirichlet Syn06','Dirichlet Syn07','Dirichlet Syn08','Dirichlet Syn09','Dirichlet Syn10'])
df_melted.columns = ['index','runs','value']

#Perform one-way anova
res = stat()
res.anova_stat(df=df_melted, res_var='value', anova_model='value ~ runs')

print("\n \n ************** ONE WAY ANOVA **************** \n")
print(res.anova_summary)

#Perform multiple pairwise comparison (Tukey's HSD)
res = stat()
res.tukey_hsd(df=df_melted, res_var='value', xfac_var='runs', anova_model='value ~ runs')

print("\n \n ************** PAIRWISE T TEST **************** \n")
print(res.tukey_summary)
```

In Fig.x, we have the boxplots of the runs obtained by using LMDirichlet and query expansions with synonyms taken from Wordnet (Tab. x).  By looking at the plot we can notice that the runs do not seem to have a normal distribution. Indeed, the first three runs present outliers and the last three runs are not balanced due to IQR located in the lower portion of the graph.

Even if the boxplots do not seem distributed as a normal distribution, since we have 50 topics for all the runs, for the Central Limit Theorem we can consider ANOVA robust to violation sof normality. Hence, we perform ANOVA Test (Tab.x) on the runs and we obtained a p-value of 1.100593 32. Since the p-value is much lower than 0.05 we reject H0 (the null hypothesis that all the means are equal) meaning that there is at least one pair of runs that are ignificantly different. Without looking at the p-value, we can affirm this, also by looking at the boxplots, since the first three are clearly different from the last three. In fact, performinga multiple pairwise comparison (Tukeys HSD Test) (Tab. x), we can notice that there are a lot of pairs that have a very low p-value. Due to this, we can say that using query expansion with synonyms we obtain a significant difference, but since the median of the run without synonyms is the highest one, it is better to not use query expansion as we already stated insection x.x

## Student Test among BEST RUN BM25 and BEST RUN LMDIRICHLET

```python
from scipy import stats

runA = df[['Dirichlet Weights Best']]
runB = df[['Bm25 Weights Best']]

t_result = stats.ttest_ind(runA,runB)

print("\n \n *************** T STUDENT TEST *************\n")
print(t_result)


print("\n \n *************** BOX PLOT *************\n")
#Boxplot
values = df[['Dirichlet Weights Best','Bm25 Weights Best']].values
plt.boxplot(values)
plt.ylabel('nDCG@5')
plt.title('Best Runs')
plt.xticks(np.arange(1,3), ['LMDirichlet Best','Bm25 Best'],rotation=90)
plt.show()
```

We compared the best run obtained using BM25 with the best run obtained using LMDirichlet. By looking at the boxplots, the runs look very different and the median of the LMDirichlet Run is higher than the BM25 one.

We then performed a T Student Test to compare the two runs and since the p-value obtained is very low (6.168497 09) and is below 0.05, we reject H0 (the null hypothesis that the two means are equal), so the run can be considered significantly different.  Moreover, since the median of the run with LMDirichlet is higher, we can confirm that LMDirichlet model is better than BM25 one for argument retrieval as we stated in section x.x
